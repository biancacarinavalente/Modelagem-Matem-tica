```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver.CreateSolver ("SCIP")
```

PROBLEMA: Quais itens devem ser colocados na  mochila de forma a maximizar o valor total dos itens escolhidos, respeitando a capacidade da mochila?

```python
# ÍNDICES DO PROBLEMA
qtd_itens = 10
```

```python
# PARÂMETROS DO PROBLEMA
peso_itens = [26, 18, 20, 35, 23, 30, 16, 14, 22, 27]
valor_itens = [132, 94, 110, 190, 120, 175, 90, 80, 115 ,160]
capacidade = 100
```

```python
# SOMANDO O PESO TOTAL DOS ITENS 
peso_total = 0
for i in range(qtd_itens):
  peso_total += peso_itens[i]

print(f' O peso total dos intens é {peso_total}')
```

```python
# SOMANDO O VALOR TOTAL DOS ITENS 
valor_total = 0
for i in range(qtd_itens):
  valor_total += valor_itens[i]

print(f' O valor total dos itens é {valor_total}')
```

```python
# VARIAVEIS DE DECISÃO
xi = list()

for i in range(qtd_itens):
  xi.append(solver.IntVar(0, 1, f"x_{i+1}"))
  
xi
```

```python
#RÓTULOS DOS ITENS

itens = list()

for i in range(qtd_itens):
  itens.append(solver.IntVar(1,1,f"x{i+1}"))

print(itens)
```

```python
# FUNÇÃO OBJETIVO

objetivo = (  
    sum(  
        valor_itens[i] * xi[i]  
        for i in range(qtd_itens)
    )
)

solver.Maximize(objetivo)
```

```python
# RESTRIÇÃO DE CAPACIDADE

solver.Add(sum(peso_itens[i] * xi[i] for i in range(qtd_itens))<= capacidade)
```

```python
# VERIFICA A SOLUÇÃO ÓTIMA

solver.set_time_limit(86400000)
status = solver.Solve()
if status == pywraplp.Solver.OPTIMAL:
    print("Solução ótima encontrada!")
    print(objetivo.solution_value())
else:
    print("Não foi encontrada solução ótima")
    print("Solução obtida:", objetivo.solution_value())
    # Melhor solução obtida
    valor_final = objetivo.solution_value()
    # Melhor solução teórica
```

```python
# VERIFICA A CAPACIDADE UTILIZADA PELA SOLUÇÃO ÓTIMA
capacidade_utilizada = sum(peso_itens[i] * xi[i] for i in range(qtd_itens))
capacidade_utilizada.solution_value()
```

```python
#IMPRIME QUAIS ITENS FORAM SELECIONADOS

for i in range(qtd_itens):
  selecionar = round(xi[i].solution_value())  #variavel selecionar vai me dizer o valor asusmido por xi
  if selecionar >0:
    print(itens[i])

```
