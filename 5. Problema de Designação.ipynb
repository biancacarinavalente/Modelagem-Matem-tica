```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver_CreateSolver ("SCIP")
```

PROBLEMA: Quais recursos vão fazer quais tarefas de forma a minimizar o custo total para executar todas as tarefas?

```python
# PARAMETROS

custo = [
    [108,312,144,270,160,300], 
    [108,208,144,360,180,270],
    [108,234,180,240,180,300],
    [144,286,180,330,240,300],
    [132,286,144,330,180,270],
    [120,312,144,270,220,300]
]

#ÍNDICES
qtd_recursos = 6
qtd_projetos = 6

#VARIAVEIS

x=list()
for i in range(qtd_recursos):
  variavel = list()
  for j in range(qtd_projetos):
    variavel.append(solver.IntVar(0,1, f'{i+1}_{j+1}'))
  x.append(variavel)
```

```python
# RESTRIÇÃO PARA RECURSOS

for i in range(qtd_recursos):
  solver.Add( 
      sum(  
          x[i][j] for j in range(qtd_projetos))==1)
```

```python
# RESTRIÇÃO PARA PROJETOS

for j in range(qtd_projetos):
  solver.Add( 
      sum(  
          x[i][j] for i in range(qtd_recursos))==1)
```

```python
# FUNÇÃO OBJETIVO

objetivo = (sum( 
    x[i][j] * custo[i][j]   
    for i in range(qtd_recursos)  
    for j in range(qtd_projetos)))

solver.Minimize(objetivo)
```

```python
# VERIFICA SOLUÇÃO ÓTIMA

status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
  print("Solução ótima encontrada")
  print(objetivo.solution_value())
else:
  print("Solução ótima não encontrada")
  print("Solução encontrada foi:",objetivo.solution_value())
```

```python
#EXIBE A SOLUÇÃO

selecionado = 0

for i in range(qtd_recursos):
  for j in range(qtd_projetos):
    selecionado = round(x[i][j].solution_value())
    if selecionado> 0:
      print(f" O recurso", i+1, "irá trabalhar no projeto", j+1)
```
