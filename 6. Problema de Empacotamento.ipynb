```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver_CreateSolver ("SCIP")
```

Problema: Quais itens colocar nas caixas de forma a minimizar a quantidade de caixas necessárias? (Os itens não podem ser fracionados)

```python
#PARÂMETROS

peso = [20,35,60,10,30,45,30,65,60,30,50,20,65,70,45]
capacidade = 100

#ÍNDICES

qtd_itens = 15
qtd_caixas = 9


# VARIAVEIS

#xij se o item i está na caixa j  

x = list()
for i in range(qtd_itens):
  variavel = list()
  for j in range(qtd_caixas):
    variavel.append(solver.IntVar(0, 1, f"{i+1}_{j+1}"))
  x.append(variavel)


#yj se a caixa j está sendo usada na solução 

y = list()
for j in range(qtd_caixas):
  y.append(solver.IntVar(0, 1, f"{j+1}"))
```

```python
#RESTRIÇÃO DE CAPACIDADE DA CAIXA

for j in range(qtd_caixas):
  solver.Add( 
      sum(  
          x[i][j] * peso[i] for i in range(qtd_itens))<=100)
```

```python
# CRIANDO UMA CAIXA COM OS ITENS QUE ESTÃO DE FORA

caixa_fora = sum(peso[i] for i in range(qtd_itens))

for i in range(qtd_itens):
  if x[i][j]==1:
    caixa_fora = caixa_fora - peso[i]
    print(caixa_fora)
```

```python
# FUNÇÃO OBJETIVO

objetivo = (sum(y[j] for j in range(qtd_caixas)))

solver.Minimize(objetivo)
```

```python
# VERIFICA SOLUÇÃO ÓTIMA

status = solver.Solve()
if status == pywraplp.Solver.OPTIMAL:
  print("Solução ótima encontrada")
  print(objetivo.solution_value())
```

```python
# IMPRIMIR RESULTADOS

for j in range(qtd_caixas):
  for i in range(qtd_itens):  
    if x[i][j].solution_value()==1:
      print(f"O item", i+1, "está na caixa", j+1)
```
